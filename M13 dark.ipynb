```python
from PIL import Image
import numpy as np
import os
import scipy
from mpldatacursor import datacursor
from scipy import ndimage
import warnings,glob
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    %matplotlib nbagg
    import matplotlib.pyplot as plt
    import matplotlib
files = glob.glob("/Users/rein/Desktop/remote_june_7/dark/resize_capt_multi_*.jpg")
print(len(files))
```

```python
s = None
for f in files[1:]:
    im = np.array(Image.open(f), dtype=np.float32)
    if s is None:
        s = im
    else:
        s += im
```

```python
vmin, vmax = np.percentile(s[:,:,2],5),np.percentile(s[:,:,2],99.9)
```

```python
fig, ax = plt.subplots(figsize=(7,5))
ax.imshow(s[:,:,2], vmin=vmin,vmax=vmax)
```
